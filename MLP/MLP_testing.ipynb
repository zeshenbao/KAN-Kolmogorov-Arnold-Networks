```python
import numpy as np
import pandas as pd
from MLP import MLP
import matplotlib.pyplot as plt
import torch
```

```python
df = pd.read_csv('/Users/hampusf/Documents/SCHOOL/KAN_proj/KAN-Kolmogorov-Arnold-Networks/datasets/data_x1+x2.csv')
```

```python
#df['x']
```

```python
X = torch.tensor(df[['x1', 'x2']].values).float()
y = torch.tensor(df['y'].values).float()
```

```python
#X = X.unsqueeze(1)
y = y.unsqueeze(1)
```

```python
input_size = X.shape[1]
hidden_sizes = [8,8,8]     # change hidden layers here
output_size = 1
n_epochs = 5000
```

```python
# Initialize and train model
model = MLP(input_size, hidden_sizes, output_size)
model.fit(X,y,n_epochs)

#print(model.forward(x[50]).detach())
#print(y[50])

# Plot results
y_approx = model.predict(X)
plt.title("y as a function of x")
plt.xlabel("x")
plt.ylabel("y")
plt.plot(X, y, 'r*', label="datapoints")  # Plots datapoints
plt.plot(X, y_approx, 'b*', label="approximation")  # Plots approximated datapoints
plt.legend()
plt.show()
```

```python
# Plot loss
plt.title("the loss as a function of number of epochs")
plt.xlabel("n_epochs")
plt.ylabel("loss")
plt.plot(model.epoch_list, model.loss_list)
```

```python
# Print parameters
for name, param in model.named_parameters():
    print(name, param.detach())
```
